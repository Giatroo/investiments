```python
from StockPlotter import MatplotStockPlotter, OneDarkPlotDecorator
from Stock import Stock

plotter = MatplotStockPlotter()
plotter = OneDarkPlotDecorator(plotter)

stocks_tickers = ['aapl34', 'mcdc34', 'abev3']

for ticker in stocks_tickers:
    stock = Stock(ticker)
    name = stock.name
    dividends = stock.get_dividends_series(period='1y')

    plotter.dividends(name, dividends)
```

```python
stocks = [Stock('itsa4'), Stock('enbr3')]
names = [stock.name for stock in stocks]
series = [stock.get_normalized_close_series(period='3mo') for stock in stocks]
```

```python
plotter.multi_lineplot(names, series, linewidths=[2], linestyles=['-', '-.'], title='Teste')
```
